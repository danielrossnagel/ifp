```python
from numpy import random
import numpy as np
import sys

x = random.rand(10, 5, 2)
xbytes = []
for idx, values in enumerate(x):
        np.append(xbytes, values.tobytes())
backtonormal = []        
for idx, values in enumerate(x):
        np.append(backtonormal, np.frombuffer(values, dtype=np.float64))

print(x)
print(xbytes)
print(backtonormal)
# As can be seen, the backtonormal form is no 100% correct. 
# Though there has to be a way so achieve this...
```
